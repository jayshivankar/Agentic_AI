```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from dotenv import load_dotenv
```

```python
class states(TypedDict):
    question : str
    answer:str
```

```python
model = ChatOpenAI()
```

```python
def get_answer(state:states) -> states:
    question = state['question']
    prompt = ChatPromptTemplate.from_template(
        "Give answer of the following question : {question}"
    )
    chain = prompt | model
    answer = chain.invoke({'question':question}).content
    state['answer'] = answer
    return state
```

```python
# define graph
graph = StateGraph(states)

# define nodes
graph.add_node('get_answer',get_answer)

# define edges
graph.add_edge(START,'get_answer')
graph.add_edge('get_answer',END)

workflow = graph.compile()
```

```python
result = workflow.invoke({'question':'Which country has highest population in the world?'})
```
